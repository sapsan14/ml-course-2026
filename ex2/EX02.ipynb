# Impordime andmeseti "tips"

Andmestik **`tips`** on väike ja tuntud näidisandmestik, mida kasutatakse laialdaselt andmeanalüüsi ja visualiseerimise õppimiseks. See kirjeldab restorani klientide arveid ja jäetud jootrahasid ning on kättesaadav Seaborn’i teegis käsu `sns.load_dataset("tips")` kaudu.

Andmestik sisaldab **244 vaatlust** (kirjet), kus iga kirje esindab ühte restoranikülastust. Iga vaatluse kohta on talletatud nii arvulised kui ka kategoorilised tunnused, mis võimaldavad mitmekülgset statistilist analüüsi.

### Tunnused andmestikus

* **`total_bill`** – arve kogusumma dollarites
* **`tip`** – jäetud jootraha dollarites
* **`sex`** – arve maksja sugu (Male, Female)
* **`smoker`** – kas klient on suitsetaja (Yes, No)
* **`day`** – nädalapäev (Thur, Fri, Sat, Sun)
* **`time`** – söögiaeg (Lunch, Dinner)
* **`size`** – inimeste arv lauas

### Kasutusvõimalused

Andmestikku kasutatakse sageli:

* jootraha ja arve summa seose analüüsimiseks
* kliendikäitumise mustrite uurimiseks
* erinevate gruppide (nt sugu, suitsetamine, nädalapäev) võrdlemiseks
* visualiseerimiste ja analüüsimeetodite õppimiseks


```python
if "google.colab" in sys.modules:
    df = sns.load_dataset("tips")
else:
    df = pd.read_csv("tips.csv")

print(df.head())
print(df.tail())
```

```python
# Väide: Meeste keskmine arve suurus on suurem kui naistel
# Joonista sobiv graafik ja vaata graafikult, kas väide peab paika


# Alusta vastust väite kinnitamise või ümberlükkamisega
# Näiteks: answer_one = "Ei, sest ..." või  answer_one = "Jah, sest ..."
sns.boxplot(data=df, x='sex', y='total_bill')
plt.show()
answer_one = "Jah, kuna meeste keskmine arve (20.74) ja mediaan on kõrgemad kui naistel (18.06)."
```

## 1\. Mida oskad öelda erinevate klasside võrdsuse kohta? Võta näiteks sugu ja suitsetajad/ mittesuitsetajad.

```python
# Joonista sobiv graafik/graafikud ja vasta paari lausega

sns.countplot(data=df, x='sex', hue='smoker')
plt.show()
answer_general_one = "Klasside vahel on tasakaalustamatus: mehi (157) on peaaegu 2 korda rohkem kui naisi (87). Mittesuitsetajaid (151) on rohkem kui suitsetajaid (93)."
```

## 2\. Mida oskad öelda tunnuste korrelatsiooni kohta?

```python
# Joonista sobiv graafik/graafikud ja vasta paari lausega

sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.show()
answer_general_two = "Kõik arvulised tunnused on positiivses korrelatsioonis. Kõige tugevam seos on arve ja jootraha (0.68) ning arve ja külaliste arvu (0.60) vahel."
```
