# Импортируем датасет "tips"

Набор данных **`tips`** — это классический пример для обучения анализу данных.

Набор содержит **244 наблюдения** (записи), где каждая запись представляет одно посещение ресторана. Для каждого наблюдения сохранены как числовые, так и категориальные признаки, что позволяет проводить разнообразный статистический анализ.

### Признаки в наборе данных

* **`total_bill`** – общая сумма чека в долларах
* **`tip`** – оставленные чаевые в долларах
* **`sex`** – пол плательщика (Male, Female)
* **`smoker`** – курит ли клиент (Yes, No)
* **`day`** – день недели (Thur, Fri, Sat, Sun)
* **`time`** – время приема пищи (Lunch, Dinner)
* **`size`** – количество людей за столом

### Возможности использования

Этот набор данных часто используется для:

* анализа связи между суммой чека и чаевыми
* изучения паттернов поведения клиентов
* сравнения различных групп (например, по полу, курению, дням недели)
* изучения методов визуализации и анализа

## 2\. Что можно сказать о корреляции признаков?

```python
# Нарисуй подходящий график и посмотри, верно ли утверждение

sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.show()
answer_general_two = "Все числовые признаки положительно коррелируют. Сильнее всего связь между счетом и чаевыми (0.68), а также счетом и количеством гостей (0.60)."
```
